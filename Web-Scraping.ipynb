## Project

Collect the the information for each category and save all information in one comma separated text file (or any other delimited file, e.g. tab separated) named rankings.txt

In addition, ask the user what category he/she would like to view rankings and then ask the user which ranking he/she would like information about. Repeat until user says he/she is done

## Solution
Below code follow this approach to get the solution:
1. Get category from user and validate it
2. Create txt file using BeautifulSoup libraries from different URLs of given website
3. Create dictionary of above data fetched from website
4. Get input from user and validate it
5. Display result from rank directory based on user input

```python
# import required libraries and functions
from bs4 import BeautifulSoup
import urllib.request

#global variables to be used for font style
esc_start_green = '\033[32;1m'
esc_start_blue = '\033[34;1m'
esc_start_bold = '\033[1m'
esc_start_red = '\033[31m'
esc_start_magenta = '\033[35;1m'
esc_start_cyan = '\033[36;1m'
esc_end = '\033[0m'

# function to create txt file from the given URLs
def create_data_file():
    file=open("rankings.txt","w+",encoding="utf-8")
    k=0
    p=0
    
    for chart in chart_types_list:
        m=0
        
        # Create dynamic URL from the chart list
        url = 'https://www.apple.com/itunes/charts/{}/'.format(chart)
        html = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(html,'lxml')
        
        # Select div which has "section-content" class from HTMl dom
        mydivs = soup.select('[class~="section-content"]')
        
        # Get all h3 tags which has name of the items
        name = mydivs[1].findAll('h3')
        
        # Get all h4 tags which has name of the items
        typeName = mydivs[1].findAll('h4')
        
        # Create rank dictionary
        rankDict=dict()
        
        for a in range(len(name)):
            nam = name[a].find('a').contents[0]
            typ = typeName[a].find('a').contents[0]
            if (a==0):
                file.write('{} \n\n'.format(chart_names[k]))
                
            # create entry of name and type in rank dictionary
            rankDict[m]="{} and {}".format(nam,typ)
            m+=1
            
            # Write name and type of items in txt file
            file.write('{}, {}, {} \n\n'.format(a+1,nam,typ))
        
        k = k+1
        rankings[chart_names[p]] = rankDict
        p+=1
        
    # Close txt file
    file.close()

cond=0

print(esc_start_blue,"What is the category that you are interested in getting data: ",esc_end)

# List for different urls
chart_types_list = ['songs','albums','free-apps','paid-apps','top-grossing-apps','paid-books','movies','tv-shows','music-videos']
# List of name of different charts
chart_names = ['Songs','Albums','Free Applications','Paid Applications','Top Grossing Applications','Books','Movies','TV Shows','Music Videos','Exit']
# List of information of different charts
chart_info = ['Song Title and Artist','Album Name and Artist','Free Application Name and Category','Paid Application Name and Category','Top Grossing Application Name and Category','Title of Book and Author','Movie Title and Category','TV Show Episode Name and TV Show Name','Music Video Name and Artist']

y=1
code = True

for typ in chart_names:
    print("({}) {}".format(y,typ))
    y+=1

# Get input from user and validate it
while(code):
    inp = input("\nEnter your choice: ")
    if(not inp.isdigit()):
         print("Input should be between 1 to 10!")
    elif(int(inp)>=1 and int(inp)<=10):
        code=False 
    else:
        print(esc_start_red,"Input should be between 1 to 10!",esc_end)

inp = int(inp)
cond = inp
flag = True

print("\n",esc_start_green,"Wait! Processing data...",esc_end)
rankings = dict()
create_data_file()

print(esc_start_green,"The data is stored in rankings.txt file",esc_end,"\n")

if(cond!=10):
     
    while(cond != 10):
        
        if(not flag):
            code = True
            # Get input from user and validate it
            while(code):
                inp = input("\nNext category you would like information about: ")
                if(not inp.isdigit()):
                     print(esc_start_red,"Input should be between 1 to 10!",esc_end)
                elif(int(inp)>=1 and int(inp)<=10):
                    code=False 
                else:
                    print(esc_start_red,"Input should be between 1 to 10!",esc_end)
            inp = int(inp)
            cond = inp
        
        if(cond!=10):
            code = True
            # Get input from user and validate it
            while(code):
                rank = input("What ranking are you interested in getting data about: ")
                if(not rank.isdigit()):
                     print(esc_start_red,"Rank should be between 1 to 100!",esc_end)
                elif(int(rank)>=1 and int(rank)<=100):
                    code=False 
                else:
                    print(esc_start_red,"Rank should be between 1 to 100!",esc_end)
            rank = int(rank)
            
            # Fetch values from the rank dictionaries
            val1 = chart_info[inp-1]
            val2 = rankings[chart_names[inp-1]][rank-1]
            
            # Print results
            print("\n",esc_start_magenta,"Your result:",esc_end)
            print("Information about ranking {}:".format(rank))
            print(esc_start_bold,"{}:".format(val1),esc_end," {}".format(val2))
            flag = False
        else:
            print(esc_start_cyan,"GoodBye!!",esc_end)
else:
    print(esc_start_cyan,"GoodBye!!",esc_end)
```
